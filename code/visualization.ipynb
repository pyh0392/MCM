```python
# 安装并导入必要的库
!pip install matplotlib seaborn pandas numpy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

为了正确显示中文，我们需要设置 matplotlib 的字体。请根据您的操作系统选择合适的字体。

```python
# 设置 matplotlib 支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

现在我们可以创建一个简单的图表示例。

```python
# 生成一些示例数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建一个简单的线图
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('这是一个简单的正弦函数图')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.grid(True)
plt.show()
```

接下来，我们使用 pandas 和 seaborn 创建一个更复杂的图表示例。

```python
# 创建一个示例 DataFrame
data = {
    '年份': [2020, 2021, 2022, 2023, 2024],
    '销售额': [100, 120, 150, 130, 170],
    '利润': [20, 25, 35, 30, 40]
}
df = pd.DataFrame(data)

# 创建一个条形图
plt.figure(figsize=(10, 6))
sns.barplot(x='年份', y='销售额', data=df)
plt.title('年度销售额')
plt.xlabel('年份')
plt.ylabel('销售额（万元）')
plt.show()
```
